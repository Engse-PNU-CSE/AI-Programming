```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# KDTree 노드 정의
class KDTreeNode:
    def __init__(self, point, left=None, right=None, axis=0):
        self.point = point
        self.left = left
        self.right = right
        self.axis = axis

# KDTree 구축
def build_kdtree(points, depth=0):
    if not points:
        return None

    k = len(points[0])
    axis = depth % k

    # 중앙값을 기준으로 정렬하여 노드를 생성
    points.sort(key=lambda x: x[axis])
    median = len(points) // 2

    return KDTreeNode(
        point=points[median],
        left=build_kdtree(points[:median], depth + 1),
        right=build_kdtree(points[median + 1:], depth + 1),
        axis=axis
    )

# 3-nearest neighbors 찾기
def find_nearest(tree, point, k=3):
    best = []
    
    def recursive_search(node):
        if node is None:
            return

        # 현재 노드와의 거리 계산
        point_distance = np.linalg.norm(np.array(point) - np.array(node.point))

        # 최근접 이웃 리스트 관리
        if len(best) < k:
            best.append((point_distance, node.point))
            best.sort(key=lambda x: x[0])
        elif point_distance < best[-1][0]:
            best[-1] = (point_distance, node.point)
            best.sort(key=lambda x: x[0])

        # 재귀적으로 검색 진행
        next_branch = None
        opposite_branch = None
        if point[node.axis] < node.point[node.axis]:
            next_branch = node.left
            opposite_branch = node.right
        else:
            next_branch = node.right
            opposite_branch = node.left

        recursive_search(next_branch)

        # 다른 분기를 검색할지 결정
        if len(best) < k or abs(point[node.axis] - node.point[node.axis]) < best[-1][0]:
            recursive_search(opposite_branch)

    recursive_search(tree)
    return [x[1] for x in best]

# 데이터 생성 및 KDTree 구축
np.random.seed(42)
points = np.random.rand(20, 3)
tree = build_kdtree(points.tolist())

# 시각화



```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for point in points:
    neighbors = find_nearest(tree, point, k=3)
    color = np.random.rand(3,)  # 각 포인트에 대해 랜덤 색상 생성
    ax.scatter(*point, color=color, s=100)  # 데이터 포인트 크기를 늘림
    for neighbor in neighbors:
        ax.scatter(*neighbor, color=color, s=50)  # 이웃 포인트
        ax.plot([point[0], neighbor[0]], [point[1], neighbor[1]], [point[2], neighbor[2]], color=color)  # 선으로 연결# 뷰 설정
ax.view_init(elev=50, azim=220)  # 높이와 아지머스 각도 설정

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D KDTree Nearest Neighbors Visualization')
plt.show()
```
